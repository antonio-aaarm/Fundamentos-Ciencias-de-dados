```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

Diagrama de dispers√£o

```python
sns.scatterplot(x = 'total_bill', y= 'tip', data = tips)
```

```python
sns.scatterplot(x = 'total_bill', y= 'tip', hue = 'day' , data = tips)
```

```python
sns.scatterplot(x = 'total_bill', y= 'tip', hue = 'day' , data = tips, size = 'size')
```

```python
sns.scatterplot(x = 'total_bill', y= 'tip', hue = 'smoker' , data = tips, palette = 'magma')
```

```python
sns.boxplot(y='tip', data = tips)
```

```python
sns.boxplot(y='tip', x= 'day',data = tips)
```

```python
sns.boxplot(y='tip', x= 'day',data = tips, hue='sex')
```
